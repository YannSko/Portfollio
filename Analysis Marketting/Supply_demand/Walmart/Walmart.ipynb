# The analysis

Group the data by Store and Date to get the total weekly sales for each store:

```python
sales_by_store = df.groupby(["Store", pd.Grouper(key="Date", freq="W-MON")])["Weekly_Sales"].sum().reset_index()
print(sales_by_store)
```

line chart of the weekly sales for each store:

```python
fig = px.line(sales_by_store, x="Date", y="Weekly_Sales", color="Store", title="Weekly Sales by Store")
fig.show()
```

We can notify a trend ( all the walmart got the same peak )

 identify stores with long checkout times:

```python
checkout_time_by_store = df.groupby("Store")["Weekly_Sales"].mean() / df.groupby("Store")["Temperature"].mean()

# Filter stores with long checkout times(more than 20m)
stores_with_long_checkout_times = checkout_time_by_store[checkout_time_by_store >= 20].index
df = df[df["Store"].isin(stores_with_long_checkout_times)]

#  average checkout time and unemployment rate for each store again  for the selected stores
checkout_time_by_store = df.groupby("Store")["Weekly_Sales"].mean() / df.groupby("Store")["Temperature"].mean()
unemployment_rate_by_store = df.groupby("Store")["Unemployment"].mean()

# df with the average checkout time and unemployment rate for each store
store_data = pd.DataFrame({"Checkout_Time": checkout_time_by_store, "Unemployment_Rate": unemployment_rate_by_store}).reset_index()
print(store_data)
#  scatter plot of the average checkout time vs. unemployment rate by store
fig = px.scatter(store_data, x="Unemployment_Rate", y="Checkout_Time", color="Store", hover_data=["Checkout_Time", "Unemployment_Rate"])
fig.update_layout(title="Average Checkout Time vs. Unemployment Rate by Store")
fig.show()
```
