```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm

data = pd.read_csv("WebsiteTraffic.csv")
print(data.head())
```

```python
data["Date"] = pd.to_datetime(data["Date"], 
                              format="%d/%m/%Y")
print(data.info())
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.plot(data["Date"], data["Views"])
plt.title("Daily Traffic of  the website Thecleverprogrammer.com")
plt.show()
```

the website traffic data is seasonal because the traffic on the website increases during the weekdays and decreases during the weekends. It is valuable to know if the dataset is seasonal or not while working on the problem of Time Series Forecasting. Below is how we can have a look at whether our dataset is stationary or seasonal:

```python
result = seasonal_decompose(data["Views"], 
                            model='multiplicative', 
                            period=  30)
fig = plt.figure()  
fig = result.plot()  
fig.set_size_inches(15, 10)
```

I will be using the Seasonal ARIMA (SARIMA) model to forecast traffic on the website. Before using the SARIMA model, it is necessary to find the p, d, and q values. 

As the data is not stationary, the value of d is 1. To find the values of p and q, we can use the autocorrelation and partial autocorrelation plots:

```python
pd.plotting.autocorrelation_plot(data["Views"])
```

```python
plot_pacf(data["Views"], lags = 100)
```

 we can train a SARIMA model for the task of website traffic forecasting:

```python
p, d, q = 5, 1, 2
model=sm.tsa.statespace.SARIMAX(data['Views'],
                                order=(p, d, q),
                                seasonal_order=(p, d, q, 12))
model=model.fit()
print(model.summary())
```

forecast traffic on the website for the next 50 days:

```python
predictions = model.predict(len(data), len(data)+50)
print(predictions)
```

```python
# how we can plot the predictions:


data["Views"].plot(legend=True, label="Training Data", 
figsize=(15, 10))
predictions.plot(legend=True, label="Predictions")
```
